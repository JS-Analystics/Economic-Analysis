```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## AS-AD Model

```python
# Reach for Output
Y = np.linspace(0, 11, 100)

# Curves
ADm = 0.8
ADy0 = 10
ASm = 0.7
ASy0 = 2
P_AD = ADy0 - ADm * Y
P_AS = ASy0 + ASm * Y

# Calculating equilibrium
Y_eq = (ADy0 - ASy0) / (ADm + ASm)
P_eq = ADy0 - ADm * Y_eq

# Plot
plt.figure(figsize=(8, 6))

plt.plot(Y, P_AD, label='AD', color='black')
plt.plot(Y, P_AS, label='SRAS', color='black')
plt.axvline(x=Y_eq, color='black', linestyle='-', label='LRAS')

# Equilibrium point
plt.scatter(Y_eq, P_eq, color='black')
#plt.text(Y_eq + 0.2, P_eq, f'E0')

# Y0 and P0
plt.plot([Y_eq, Y_eq], [0, P_eq], color='black', linestyle='--')
plt.xticks([Y_eq], ['Y*'])
plt.plot([0, Y_eq], [P_eq, P_eq], color='black', linestyle='--')
plt.yticks([P_eq], ['P*'])

# Axis labels
plt.xlabel('Real GDP (Y)')
plt.ylabel('Price Level (P)')
plt.title('AS-AD Model')

# Styling
plt.grid(False)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# Choosing position for labels
x_label = 10

# Determining respective y values for labels
y_AD_label = ADy0 + 0.2 - ADm * x_label
y_AS_label = ASy0 + 0.5 + ASm * x_label

# labeling the curves
plt.text(x_label, y_AD_label, 'AD', fontsize=11, ha='left', va='bottom')
plt.text(x_label, y_AS_label, 'SRAS', fontsize=11, ha='left', va='bottom')
plt.text(Y_eq + 0.1, max(P_AS), 'LRAS', fontsize=11, va='bottom')


plt.show()
```

## New Keynesian Model Production Function

```python
# Input
N = np.linspace(0.1, 11, 100)  # 0 vermeiden wegen Potenz

#Setting Values for Production Function
z = 1
K = 10

# Cobb-Douglas Production Function: zF(K,N) = z * K^0.3 * N^0.7, z = Total Factor Productivity, K = Capital, N = Labor
# Assumption about exponentials in Production Function according to empirical source in Macroeconomics by Stephen D. Williamson 6th Edition page 152
prod = z * K**0.3 * N**0.7

# Example (e.g. N* = 5)
N_star = 5
Y_star = z * K**0.3 * N_star**0.7

# Plot
plt.figure(figsize=(8, 6))

# Curve
plt.plot(N, prod, color='black')

# Example point
plt.scatter(N_star, Y_star, color='black')

# Equilibrium lines (only to the axis)
plt.plot([N_star, N_star], [0, Y_star], linestyle='--', color='black')
plt.plot([0, N_star], [Y_star, Y_star], linestyle='--', color='black')

# Equilibrium lables
plt.xticks([N_star], ['N*'])
plt.yticks([Y_star], ['Y*'])

# Lables
plt.xlabel('Labor (N)')
plt.ylabel('Output (Y)')
plt.title('Cobb-Douglas Production Function')

# Styling
plt.grid(False)
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.show()
```

# Second Round Effect

## SR - Labor Market

```python
# Function Values
Y = np.linspace(0, 11, 100)

# Curves
Ndm = 0.8
Ndy0 = 10
Nsm = 0.7
Nsy0 = 2
P_Nd = Ndy0 - Ndm * Y
P_Nd2 = (Ndy0-2) - Ndm * Y
P_Ns = Nsy0 + Nsm * Y
P_Nd3 = (Ndy0-1) - Ndm * Y
P_Ns3 = (Nsy0+0.3) + Nsm * Y

# Calculating equilibrium
N_eq = (Ndy0 - Nsy0) / (Ndm + Nsm)
w_eq = Nsy0 + Nsm * N_eq
N2 = ((Ndy0 - 2) - w_eq) / Ndm
N3 = ((Ndy0 - 2) - Nsy0) / (Ndm + Nsm)
w3 = Nsy0 + Nsm * N3
N4 = ((Ndy0 - 1) - (Nsy0 + 0.3)) / (Ndm + Nsm)
w4 = (Nsy0 + 0.3) + Nsm * N4

# Plot
plt.figure(figsize=(8, 6))

plt.plot(Y, P_Nd, label='Nd', color='black', alpha=0.2)
plt.plot(Y, P_Nd2, label='Nd2', color='red', alpha=0.2)
plt.plot(Y, P_Ns, label='Ns', color='black', alpha=0.2)
plt.plot(Y, P_Nd3, label='Nd3', color='orange')
plt.plot(Y, P_Ns3, label='Ns3', color='orange')

# Y0 and P0
plt.plot([N_eq, N_eq], [0, w_eq], color='black', linestyle='--', alpha=0.2)
plt.plot([0, N_eq], [w_eq, w_eq], color='black', linestyle='--', alpha=0.2)
plt.yticks([w_eq], ['w1'])
plt.plot([N2, N2], [0, w_eq], color='red', linestyle='--', alpha=0.2)
plt.plot([N3, N3], [0, w3], color='orange', linestyle='--', alpha=0.2)
plt.plot([0, N3], [w3, w3], color='orange', linestyle='--', alpha=0.2)
plt.plot([N4, N4], [0, w4], color='orange', linestyle='--')
plt.plot([0, N4], [w4, w4], color='orange', linestyle='--')

# Lableling Axis
ax = plt.gca()
ax.set_xticks([N_eq, N2, N3, N4])
ax.set_xticklabels(['N1', 'N2', 'N2.1', 'N3'])
ax = plt.gca()
ax.set_yticks([w_eq, w3, w4])
ax.set_yticklabels(['w1', 'w2.1', 'w3'])

# Setting Colours
xticklabels = ax.get_xticklabels()
xticklabels[0].set_color('black')
xticklabels[1].set_color('red')
xticklabels[2].set_color('orange')
xticklabels[3].set_color('orange')

yticklabels = ax.get_yticklabels()
yticklabels[0].set_color('black')
yticklabels[1].set_color('orange')
yticklabels[2].set_color('orange')

# Axis labels
plt.xlabel('Labor (N)')
plt.ylabel('Real Wage (w)')
plt.title('Labor Market')

# Styling
plt.grid(False)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# Choosing position for labels
x_label = 10

# Determining respective y values for labels
y_Nd_label = Ndy0 + 0.2 - Ndm * x_label
y_Ns_label = Nsy0 + 0.5 + Nsm * x_label
y_Nd2_label = (Ndy0-2) + 0.2 - Ndm * x_label

# labeling the curves
plt.text(x_label, y_Nd_label, 'Nd1', fontsize=11, ha='left', va='bottom', alpha=0.2)
plt.text(x_label, y_Ns_label, 'Ns1', fontsize=11, ha='left', va='bottom', alpha=0.2)
plt.text(x_label, y_Nd2_label, 'Nd2', fontsize=11, ha='left', va='bottom', color='red', alpha=0.2)
plt.text(x_label, y_Nd2_label + 1, 'Nd3', fontsize=11, ha='left', va='bottom', color='orange')
plt.text(x_label, y_Ns_label + 0.5, 'Ns3', fontsize=11, ha='left', va='bottom', color='orange')


plt.show()
```
